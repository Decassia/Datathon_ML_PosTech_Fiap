```python
# Histograma
def gerar_graficos_inde_ano(df_inde, ano_especifico):
    # Garantir que a coluna INDE seja numérica
    df_inde['INDE'] = pd.to_numeric(df_inde['INDE'], errors='coerce')

    # Filtrar o DataFrame pelo ano específico
    df_ano = df_inde[df_inde['ANO'] == ano_especifico]
    inde = df_ano['INDE'].dropna()

    # Calcular a média de idade
    media_inde = inde.mean()

    # Criar a figura
    fig = make_subplots(rows=1, cols=2, subplot_titles=(f'Distribuição - {ano_especifico}', 'Boxplot'),
                        column_widths=[0.6, 0.4])

    # Subplot 1: Histograma
    fig.add_trace(go.Histogram(
        x=inde,
        nbinsx=10,
        opacity=0.7,
        name=f'Histograma {ano_especifico}',
        marker=dict(color='blue', line=dict(color='lightblue', width=1)),
    ), row=1, col=1)



    # Adicionar a curva KDE
    kde = gaussian_kde(inde)
    x_vals = np.linspace(min(inde), max(inde), 100)
    kde_vals = kde(x_vals)

    fig.add_trace(go.Scatter(
        x=x_vals,
        y=kde_vals * len(inde),  # Normalizar a densidade
        mode='lines',
        line=dict(color='red'),
        name='KDE'
    ), row=1, col=1)

    # Subplot 2: Boxplot com a média destacada
    fig.add_trace(go.Box(
        y=inde,
        boxmean=True,  # Exibir a média no boxplot
        boxpoints=False,
        marker_color='#D2691E',
        name='Boxplot'
    ), row=1, col=2)

    # Adicionar o valor da média no gráfico
    fig.add_annotation(
        x=1, y=media_inde, xref="x2", yref="y2",
        text=f"Média: {media_inde:.1f}",
        showarrow=False, font=dict(size=12, color="black"),
        align="center", bordercolor="black", borderwidth=1, borderpad=4,
        bgcolor="white", opacity=0.8
    )

    # Ajustar layout
    fig.update_layout(
        title_text=f'Análise de INDE - {ano_especifico}',
        xaxis_title='Valor',
        yaxis_title='Frequência',
        template='plotly_white',
        showlegend=True,
        height=800,
        width=1200
    )


    fig.show()



```

```python
# Função que compara a media do INDE dos anos de 22,23,24
def box_plot_comparative(data_anos, labels, title):
    #fig = make_subplots(rows=1, cols=len(data_anos), subplot_titles=labels) # Create figure with subplots
    fig = make_subplots(rows=1, cols=2, subplot_titles=('', ''),
                        column_widths=[0.6, 0.4])


    # Iterando sobre os anos e seus respectivos dados
    for i, data in enumerate(data_anos):
        # Convertendo a coluna 'INDE' para numérica
        # Usando 'coerce' para transformar valores inválidos em NaN
        data['INDE'] = pd.to_numeric(data['INDE'], errors='coerce')

        # Criando o box plot para cada ano
        fig.add_trace(go.Box(
            y=data['INDE'],  # Dados do ano i
            boxmean=True,
            boxpoints=False,  # Não mostrar os pontos (bolinhas)
            marker_color='#D2691E',  # Cor da caixa
            #line=dict(color="black"),  # Cor das bordas da caixa
            name=labels[i]  # Nome de cada boxplot (ano)
        ))


        # Adicionando a linha de média para cada ano
        mean_value = np.mean(data['INDE'])  # Calculando a média dos dados
        fig.add_trace(go.Scatter(
            x=[i-0.2, i+0.2],  # Colocando a linha no centro de cada boxplot
            y=[mean_value, mean_value],
            mode="lines",
            line=dict(color="red", width=2, dash='dash'),
            name=f'Média {labels[i]}: {mean_value:.1f}'
        ))

    # Customizando o layout
    fig.update_layout(
        title=title,
        yaxis_title='Valores',
        xaxis_title='Anos',
        template='plotly_white',
        boxgap=0.30,       # Ajusta o espaçamento entre as caixas
        boxgroupgap=0.3,
        showlegend=True,  # Exibir a legenda para as médias
        height=600, width=1000  # Tamanho da figura
    )



    fig.show()

```

```python
#Graficos

def plot_exact_counter(size, x, y, df):
    # Definindo as cores: ametista, ágata, quartzo, topázio
    colors = ['#9966CC', '#D2691E', '#D3D3D3', '#007FFF']

    # Criando o gráfico de barras com plotly.graph_objects
    fig = go.Figure(go.Bar(
        x=y.index,  # Usando o índice da Series (categorias)
        y=y.values,  # Usando os valores (contagens)
        marker_color=colors[:len(y)],  # Aplicando as cores personalizadas
        text=y.values,  # Exibindo os valores nas barras
        textposition='outside'  # Posicionando os valores fora das barras
    ))

    # Customizando o layout do gráfico
    fig.update_layout(
        title=f'Gráfico de Barras: {x}',  # Título dinâmico com o nome de x
        xaxis_title='Categorias',
        yaxis_title='Contagem',
        template='plotly_white',  # Estilo de fundo branco
        plot_bgcolor='rgba(0,0,0,0)',  # Fundo transparente
        showlegend=False,  # Não mostrar legenda
        height=size[1] * 100,  # Ajustando altura conforme o tamanho desejado
        width=size[0] * 100  # Ajustando largura conforme o tamanho desejado
    )

    # Personalizando o eixo y para exibir apenas inteiros
    fig.update_yaxes(tickformat='d')



    # Exibir o gráfico
    fig.show()
```

## Verficando a distribuiçao dos alunos por faixa etaria

```python
# Funçao que processa a distribuiCao da idade por ano

def processar_dist_idade_todos_anos(df_idade):
    # Garantir que a coluna IDADE seja numérica
    df_idade['IDADE'] = pd.to_numeric(df_idade['IDADE'], errors='coerce')

    # Obter todos os anos únicos
    anos = sorted(df_idade['ANO'].unique())

    # Criar DataFrames vazios para armazenar as distribuições e estatísticas descritivas
    distribuições_idade = pd.DataFrame()
    estatisticas_idade = pd.DataFrame()

    # Iterar sobre cada ano e processar as idades
    for ano in anos:
        df_ano = df_idade[df_idade['ANO'] == ano]  # Filtrar o DataFrame por ano

        # Distribuição de Idades
        dist_idade = df_ano['IDADE'].value_counts().sort_index()  # Contagem de valores
        distribuições_idade[f'ANO {ano}'] = dist_idade  # Adiciona ao DataFrame
        total_alunos =dist_idade.sum()


        # Estatísticas descritivas
        estatisticas = df_ano['IDADE'].describe()  # Estatísticas descritivas
        estatisticas_idade[f'ANO {ano}'] = estatisticas  # Adiciona ao DataFrame

    # Preencher valores NaN com zeros (opcional)
    distribuições_idade = distribuições_idade.fillna(0)


     # Adicionar uma coluna para a soma total de alunos em todos os anos
    distribuições_idade['TOTAL'] = distribuições_idade.sum(axis=1)

    # Exibir os resultados
    print("\nDistribuições de Idade por Ano com Soma Total de Alunos:\n")
    display(distribuições_idade)


    print("\nEstatísticas Descritivas por Ano:\n")
    display(estatisticas_idade)
```

Observamos que a idade dos alunos varia entre 7 e 20 anos, com as maiores concentrações em faixas específicas, como as idades de 13 e 12 anos.

- Nas estatísticas descritivas por ano, observamos que foi mantido a idade média dos alunos, de 12 anos em 2022 e para os anos seguintes. A variação (desvio padrão) diminui ao longo dos anos, indicando uma concentração maior de alunos em torno de idades específicas, especialmente em 2022, quando o desvio padrão é de 1,36. Os valores mínimos e máximos para cada ano também mostram que os alunos mais novos e mais velhos estão progredindo em idade com o passar do tempo.

```python
# Função para calcular os dados para o gráfico de radar
def calcular_dados_aluno(df, aluno, ano):
    df_aluno = df[(df['NOME'] == aluno) & (df['ANO'] == ano)].drop(columns=['NOME', 'ANO'])

    if df_aluno.empty:
        return pd.DataFrame()  # Retorna um DataFrame vazio se não houver dados para o aluno/ano especificado

    # Transpor e preparar o DataFrame para o gráfico
    df_transposto = df_aluno.T.reset_index()
    print('\n')
    df_transposto.columns = ['Métricas', 'Valores']

    df_transposto = df_transposto.dropna()  # Remove possíveis NaNs

    return df_transposto
```

```python
# Função para plotar o gráfico de radar
def radar_chart(df, title):
    labels = df['Métricas'].values
    values = df['Valores'].values

    # Adicionar o primeiro valor ao final para fechar o gráfico
    values = np.concatenate((values, [values[0]]))

    # Criar ângulos igualmente espaçados para cada métrica
    angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
    angles += angles[:1]

    # Criar o gráfico de radar
    fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

    # Plotar os dados
    ax.fill(angles, values, color='orange', alpha=0.25)
    ax.plot(angles, values, color='#D2691E', linewidth=2)

    # Adicionar os rótulos das métricas
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels)

    # Adicionar os valores ao lado de cada ponto
    for i, (angle, value) in enumerate(zip(angles, values)):
        if i < len(labels):  # Evitar colocar o valor duas vezes no fechamento do gráfico
            ax.text(angle, value + 0.5, f'{value:.1f}', ha='center', va='center', fontsize=10, color='blue')



    # Ocultar os números dos eixos
    ax.set_yticklabels([])

    # Título do gráfico
    plt.title(title, size=15, color='black', y=1.1)

    plt.show()
```

```python
# Widgets para seleção
aluno_dropdown = widgets.Dropdown(
    options=df_alunos['NOME'].unique(),
    description='Aluno:',
    value=df_alunos['NOME'].unique()[0]
)

ano_dropdown = widgets.Dropdown(
    options=df_alunos['ANO'].unique(),
    description='Ano:',
    value=df_alunos['ANO'].unique()[0]
)

# Função para atualizar o gráfico com base nas seleções
def atualizar_grafico(aluno, ano):
    df_aluno = calcular_dados_aluno(df_alunos, aluno, ano)
    if df_aluno.empty:
        print(f"\nSem dados para o aluno {aluno} no ano {ano}.\n")
        return

    print(df_aluno)
    radar_chart(df_aluno, title=f'Desempenho de {aluno} no ano {ano}')

# Criar widget interativo
interativo = widgets.interactive(atualizar_grafico, aluno=aluno_dropdown, ano=ano_dropdown)

# Exibir widget
print('\n')
display(interativo)
print('\n')

```
